
# üß† Step 1: Inspecting the Data

Functions to explore:

## Head()

Esta funci√≥n ofrece el n√∫mero de resultados que se le env√≠e como argumento siempre empezando por arriba. Por ejemplo, si le env√≠as el n√∫mero 10, te ofrecer√° los primeros 10 resultados. (Por defecto muestra los primeros 5 resultados).

```python

# Ejemplo 1: por defecto, muestra los primeros 5 resultados.

print(df.head())

# Example 2: le pasamos el n√∫mero 10 como par√°metro, 
# por lo tanto, muestra los primeros 10 resultados.

df.head(10)
```

## Tail()

Al contrario de head(), que te muestra los resultados desde el principio, tail() te ofrece los resultados a partir del final del dataframe. Es decir, si le pasamos el mismo n√∫mero que en ejemplo anterior, tail nos devolver√° los √∫ltimos 10 resultados (tambi√©n devuelve 5 por defecto).

```python
# Ejemplo 1: por defecto, muestra los 5 √∫ltimos.

print(df.tail())

# Ejemplo 2: devuelve los √∫ltimos 10 resultados del dataframe.

df.tail(10)
```

## Info()

Info() nos da informaci√≥n sobre los tipos de datos, el numero de no nulos y el espacio en memoria. Tambi√©n el n√∫mero total de columnas y otros tipos de informaci√≥n sobre el dataset.

```python
# Ejemplo 1: muestra los datos del dataframe donde se ejecuta. Muestra el total de entradas y columnas,
# posteriomente te muestra las cabeceras de las columnas con la informaci√≥n de cada una de ellas. Tambi√©n
# muestra el n√∫mero de diferentes tipos de datos y el espacio de memoria ocupado por el dataframe.

print(df.info())

# Ejemplo 2: muestra la informaci√≥n detallada de la columna seleccionada.

print(df[["Cabin"]].info())
```

## Describe()

Describe() nos proporciona informaci√≥n estad√≠stica sobre los datos, ademas de los cuartiles, la media y la desviaci√≥n t√≠pica. Tambi√©n nos da los maximos, m√≠nimos y n√∫mero de filas.

```python
# Ejemplo 1: muestra informaci√≥n estad√≠stica de las columnas del dataframe.

print(df.describe())

# Ejemplo 2: tambi√©n se puede sacar esa informaci√≥n seleccionando la columna a inspeccionar.

print("\n\n",df["Fare_per_Age"].describe())
```

## Shape

Shape devuelve una tupla con el n√∫mero de filas y columnas. Se usa para inspeccionar r√°pidamente el tama√±o de un conjunto de datos sin imprimirlo completo.

```python
# Ejemplo:

print(df.shape)
```

## Columns

Columns devuelve los nombres de las columnas. Se usa para inspeccionar, renombrar o manipular las columnas de un dataset.

```python
# Ejemplo 1: te devuelve la lista de todas las columnas.

print(df.columns)

# Ejemplo 2: Puedes referir a una columna en espec√≠fico seg√∫n su index.

print(df.columns[0])

```


# üîç Step 2: Selecting Data

Functions to explore:
- df["column"]
- df[["col1", "col2"]]
- df.loc[]
- df.iloc[]


## ["column"]

Devuelve una serie que contiene los valores de la columna a la que te refieras. Es una de las formas m√°s comunes y directas de acceder a los datos de una columna dentro de un DataFrame.

```python
# Ejemplo 1: Todos los registros de la columna "Name"
print(df["Name"])

# Ejemplo 2: Todos los registros de "Age" + 10
print(df["Age"]+10)
```

## [["col1","col2"]]

Devuelve un dataframe con las columnas especificadas. Se usa para seleccionar y trabajar sobre las columnas deseadas dentro de un dataframe m√°s extenso.

```python
# Ejemplo 1: Lista de "Name" y "Age".
print(df[["Name","Age"]])

# Ejemplo 2: Lista de "Name" y "Age" sumandole 10 a todas estas.
nombreEdad = df[["Name","Age"]]
nombreEdad["Age"] = nombreEdad["Age"]+10
print(nombreEdad)
```

## Loc[]

Permite seleccionar filas y columnas por sus etiquetas (no por posici√≥n num√©rica). Puede usarse para filtrar filas por condiciones, seleccionar columnas espec√≠ficas o asignar valores.

```python
# Ejemplo 1: primeros 10 registros de "Name" y "Age"
print(df.loc[0:10, ["Name","Age"]])

# Ejemplo 2: todos los registros de "Name" y "Age" de los menores de 18 a√±os
print(df.loc[df["Age"] < 18, ["Name","Age"]])
```

## Iloc[]

Permite seleccionar filas y columnas por posici√≥n entera (√≠ndices num√©ricos). A diferencia de loc[], que trabaja con etiquetas, iloc[] siempre utiliza posiciones basadas en cero.

```python
# Ejemplo 1: filas y columnas de la 5 a la 10
print(df.iloc[5:10, 5:10])

# Ejemplo 2: primeros dos registros mostrados al rev√©s
print(df.iloc[:2,::-1])
```


# üîé Step 3: Filtering Rows

Functions to explore:
- df[df["Age"] > 30]
- df.query("Sex == 'female' and Survived == 1")


## ["Age" > 30]

Genera una Serie de valores True/False. Luego, df[...] devuelve √∫nicamente las filas donde la condici√≥n es True. Es el mecanismo est√°ndar en pandas para filtrar registros seg√∫n una condici√≥n.

```python
# Ejemplo 1: todos los registros de los mayores de 30 a√±os
print(df[df["Age"] > 30])

# Ejemplo 2: los campos de "Name", "Age" y "Fare" de los menores de 18 a√±os
df[df["Age"] < 18][["Name","Age","Fare"]]
```

## Query()

Permite filtrar filas de un DataFrame utilizando una sintaxis similar a expresiones l√≥gicas en SQL, lo que mejora la legibilidad frente a filtros largos con corchetes. Dentro del m√©todo se escribe una cadena con la condici√≥n, usando directamente los nombres de las columnas.

```python
# Ejemplo 1: todos los registros de los menores de 18 a√±os
df.query("Age < 18")

# Ejemplo 2: menores de 18 a√±os que han sobrevivido
df.query("Age < 18 and Survived == 1")
```


# üßπ Step 4: Handling Missing Data

Functions to explore:
- df.isna()
- df.isna().sum()
- df.dropna()
- df.fillna()


## Isna()

Detecta valores nulos en el DataFrame y devuelve un DataFrame de booleanos (True si el valor es nulo, False si no lo es).

```python
# Ejemplo 1: comprobacion de nulos por columna de los primeros 5 registros
print(df.isna().head())

# Ejemplo 2: comprobacion de nulos en una columna especifica en los primeros 5 registros
print(df["Cabin"].isna().head())
```

## Isna().sum()

Combina isna() con sum() para contar cu√°ntos valores nulos hay en cada columna del DataFrame.

```python
# Ejemplo 1: suma de los valores nulos por columna
print(df.isna().sum())

# Ejemplo 2: Numero de nulos de una columna espec√≠fica
print(f"N√∫mero de nulos de Age --> {df["Age"].isna().sum()}")
```

## Dropna()

Elimina filas (o columnas) que contienen valores nulos (NaN).

```python
# Ejemplo 1: primeros 5 registros que no contienen nulos.
print(df.dropna().head())

# Ejemplo 2: primeros 5 registros sin las columnas que contenian alg√∫n nulo en general.
print(df.dropna(axis=1).head())
```

## Fillna()

Rellena los valores nulos (NaN) con un valor espec√≠fico o con alg√∫n m√©todo como promedio, adelante, atr√°s, etc.

```python
# Ejemplo 1: relleno de 0 en los nulos de age
df["Age"].fillna(0)
print(f"N√∫mero de nulos en Age despu√©s del fillna() --> {df["Age"].isna().sum()}")
# Ejemplo 2: rellenado de nulos en cabin con la palabra unknown
df["Cabin"] = df["Cabin"].fillna("Unknown")
print(df["Cabin"].head())
print(df.isna().sum())
```


# üìä Step 5: Grouping and Aggregating

Functions to explore:
- df.groupby("Sex")["Survived"].mean()
- df["Fare"].mean()
- df["Age"].median()


## Groupby()[].mean()

Agrupa los datos por una o m√°s columnas y calcula el promedio de otra columna num√©rica.

```python
# Ejemplo 1: precio medio de las clases del titanic.
print(df.groupby(["Pclass"])["Fare"].mean())

# Ejemplo 2: precio medio de las clases del titanic por g√©nero.
print(df.groupby(["Pclass","Sex"])["Fare"].mean())

```

## [].mean()

Calcula la media de los datos de la columna especificada siempre que sean datos num√©ricos.

```python
# Ejemplo 1: Media de precio del ticket del titanic.
print(f" Precio medio: {df["Fare"].mean()}")

# Ejemplo 2: Edad promedia de los tripulantes del titanic.
print(f"Edad promedia: {df["Age"].mean()}")

```

## [].median()

Saca la mediana de los datos de la columna especificada siempre que sean datos num√©ricos.

```python
# Ejemplo 1: Mediana de precio del ticket del titanic.
print(f" Precio mediano: {df["Fare"].median()}")

# Ejemplo 2: Edad mediana de los tripulantes del titanic.
print(f"Edad mediana: {df["Age"].median()}")
```


# üìà Step 6: Sorting and Counting

Functions to explore:
- df.sort_values("Age")
- df["Sex"].unique()
- df["Pclass"].value_counts()


## sort_values()

Ordena el DataFrame seg√∫n los valores de una o m√°s columnas.

```python
# Ejemplo 1: tabla ordenada de menor a mayor seg√∫n la edad.
print(df.sort_values("Age").head())

# Ejemplo 2: tabla ordenada de menor a mayor seg√∫n la edad y el precio pagado.
print(df.sort_values(["Age", "Fare"]).head())
```

## unique()

Devuelve un array con los valores √∫nicos de una columna quitando los duplicados si existen.

```python
# Ejemplo 1: datos recogidos dentro de "Pclass" en formato array.
print(df["Pclass"].unique())

# Ejemplo 2: n√∫mero de registros recogidos dentro de "Pclass".
print(len(df["Pclass"].unique()))
```

## value_counts()

Cuenta cu√°ntas veces aparece cada valor √∫nico en una columna.

```python
# Ejemplo 1: n√∫mero de personas por clase 
print(df["Pclass"].value_counts())

# Ejemplo 2: n√∫mero de personas por clase normalizado
print(df["Pclass"].value_counts(normalize=True))
```

## Assign()

Crea nuevas columnas o modifica existentes sin alterar el DataFrame original, devolviendo una copia con los cambios.

```python
# Ejemplo 1: nuevo dataframe con el a√±o de nacimiento de los tripulantes 
# (suponiendo que el dataframe es actual)
print(df.assign(YearOfBirth = abs(df["Age"] - 2025)).head())

# Ejemplo 2: nuevo dataframe con una columna que guarda la edad dentro de 10 a√±os.
print(df.assign(Age_in_10_Years = df["Age"] + 10).head())
```

## Apply()

Aplica una funci√≥n (propia o de Python) a cada elemento, fila o columna del DataFrame o Series.

```python
# Ejemplo 1: suma de las edades de los tripulantes
dfnn = df.fillna({"Age":0})
print(dfnn[["Age"]].apply(sum))

# Ejemplo 2: edades de los tripulantes * 2.
print(df[["Age"]].apply(lambda x: x*2).head())
```

## ["new_col"] = ...

Crea o modifica una columna en el DataFrame asign√°ndole un valor, lista o c√°lculo.

```python
# Ejemplo 1: a√±adir columna booleana de adultos.
df["Adulto"] = df["Age"] >= 18
print(df.head())

# Ejemplo 2: a√±adir columna sumandole 10 a las edades en "Age".
df["Age_in_10_Years"] = df["Age"] + 10
print(df)
```
